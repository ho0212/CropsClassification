```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import shutil
import cv2
from sklearn.model_selection import train_test_split
from PIL import Image
import tensorflow as tf
```

```python
import pandas as pd
submit=pd.read_csv('F:\\AICUP\\upload\\TTA_800(0.756)_1.csv')
submit
```

```python
#抓出預測為以下類別的圖片
bambooshoot=list()
betel=list()
for i in range(len(submit['label'])):
    if submit['label'][i]=='bambooshoots':
        bambooshoot.append(submit['filename'][i])
    elif submit['label'][i]=='betel':
        betel.append(submit['filename'][i])
print(len(bambooshoot))
print(len(betel))
```

```python
img_path="F:/AICUP/public_800/"
for i in range(len(bambooshoot)):
    tmp=img_path+bambooshoot[i]
    bambooshoot[i]=tmp
for i in range(len(betel)):
    tmp=img_path+betel[i]
    betel[i]=tmp
    
repredict=bambooshoot+betel
len(repredict)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.xception import Xception,preprocess_input
```

```python
model=keras.models.load_model('F:\\AICUP\\trainset_400\\model-logs\\bambooshoots_betel.h5')
```

```python
img_shape=(400,400)
result=list()
i=1
for img in repredict:
    img=Image.open(img)
    img=img.resize(img_shape)
    input_arr = tf.keras.preprocessing.image.img_to_array(img)
    input_arr = preprocess_input( np.array([input_arr]) )
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)
    print(i)
    i+=1
```

```python
for i in range(len(result)):
    if result[i]==0:
        result[i]='bambooshoots'
    elif result[i]==1:
        result[i]='betel'
result
```

```python
filelist=list()
for file in repredict:
    file=file.split('/')[-1]
    print(file)
    filelist.append(file)
```

```python
#confirm the length of filelist and result
print(len(filelist))
print(len(result))
```

```python
for i in range(len(filelist)):
    file=filelist[i]
    submit['label'][submit['filename']==file]=result[i]
submit
```

```python
submit.to_csv('F:\\AICUP\\upload\\specific_classes(0.756).csv',index=False)# index=False去掉索引
```

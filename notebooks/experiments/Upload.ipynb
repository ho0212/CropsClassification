```python
import tensorflow as tf
model = tf.keras.models.load_model('D:\\AICUP\\trainset_400\\model-logs\\xception(0.786).h5') #  
model.summary()
```

```python
import cv2
import os
i=0
data = [] # 
# labels = []# Label
files = [] # 
for path in jpglist:
    i+=1
    im1 = cv2.imread(path)
    im1=cv2.resize(im1,(128,128))
#     print(path)
    file = os.path.split(path)[-1] # 
#     print(file)
    data.append(im1) #
    files.append(file) #
    print(i)
data = np.array(data)/255.0 # /255.0  z-score
```

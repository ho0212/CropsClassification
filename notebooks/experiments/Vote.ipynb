```python
import pandas as pd
import numpy as np
org_crop=pd.read_csv('F:\\AICUP\\upload_private\\org_crop_800.csv')
org=pd.read_csv('F:\\AICUP\\upload_private\\org_800.csv')
crop=pd.read_csv('F:\\AICUP\\upload_private\\crop_800.csv')
```

```python
df=org_crop
df.rename(columns={'label':'org_crop'},inplace=True)
df['org']=org['label']
df['crop']=crop['label']
df
```

```python
def unique(list1):
    x = np.array(list1)
    return x
```

```python
result=list()
for i in range(len(df)):
    answer_list=list()
    ans_org_crop=df['org_crop'][i]
    answer_list.append(ans_org_crop)
    ans_org=df['org'][i]
    answer_list.append(ans_org)
    ans_crop=df['crop'][i]
    answer_list.append(ans_crop)
    predictions=max(answer_list,key=answer_list.count)
    result.append(predictions)
```

```python
df['vote']=result
df
```

```python
submit=pd.DataFrame(df['filename'])
submit['label']=df['vote']
submit
```

```python
import tensorflow as tf
from tensorflow import keras
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.densenet import DenseNet201,preprocess_input
```

```python
model=keras.models.load_model('F:\\AICUP\\trainset_400\\model-logs\\Densenet_org(0.808).h5') #讀取訓練完成的模型
model.summary()
```

```python
def flipimg(img_path,img_shape):
    img=Image.open(img_path)
    img=img.resize(img_shape)
    lr=img.transpose(Image.FLIP_LEFT_RIGHT)
    ud=img.transpose(Image.FLIP_TOP_BOTTOM)
    rotate90=img.transpose(Image.ROTATE_90)
    rotate270=img.transpose(Image.ROTATE_270)
    return img,lr,ud,rotate90,rotate270
```

```python
def contrast(img_path):
    img=Image.open(img_path)
    img=img.resize(img_shape)
    img=np.array(img)
    contrast = 200
    brightness = 0
    output = img * (contrast/127 + 1) - contrast + brightness 
    output = np.clip(output, 0, 255)
    output = np.uint8(output)
    output=Image.fromarray(output)
    return output
```

```python
def TTAPredict(img,lr,ud,rotate90,rotate270,contrast_img):
    result=list()
    #original img
    input_arr = tf.keras.preprocessing.image.img_to_array(img)
    input_arr = preprocess_input( np.array([input_arr]) )/255
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)
    
    #horizontal_flip
    input_arr = tf.keras.preprocessing.image.img_to_array(lr)
    input_arr = preprocess_input( np.array([input_arr]) )/255
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)
    
    #vertical_flip
    input_arr = tf.keras.preprocessing.image.img_to_array(ud)
    input_arr = preprocess_input( np.array([input_arr]) )/255
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)

    #rotate90
    input_arr = tf.keras.preprocessing.image.img_to_array(rotate90)
    input_arr = preprocess_input( np.array([input_arr]) )/255
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)

    #rotate270
    input_arr = tf.keras.preprocessing.image.img_to_array(rotate270)
    input_arr = preprocess_input( np.array([input_arr]) )/255
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)
    
    #contrast
    input_arr = tf.keras.preprocessing.image.img_to_array(contrast_img)
    input_arr = preprocess_input( np.array([input_arr]) )/255
    predictions = model.predict(input_arr).argmax()
    result.append(predictions)
    
    answer=max(result,key=result.count)
    return answer
```

```python
answer_list=list()
i=1
for jpg in jpglist:
    a,b,c,d,e=flipimg(jpg,img_shape)
    f=contrast(jpg)
    answer=TTAPredict(a,b,c,d,e,f)
    answer_list.append(answer)
    print(i)
    i+=1
```

Vote(執行該部分可將多個版本的答案進行投票選出最後的答案)

```python
#輸入由TTA Predict所得出的答案
import pandas as pd
import numpy as np
org_crop=pd.read_csv('F:\\AICUP\\upload_private\\org_crop_800.csv')
org=pd.read_csv('F:\\AICUP\\upload_private\\org_800.csv')
crop=pd.read_csv('F:\\AICUP\\upload_private\\crop_800.csv')
```

```python
df=org_crop
df.rename(columns={'label':'org_crop'},inplace=True)
df['org']=org['label']
df['crop']=crop['label']
df
```

```python
def unique(list1):
    x = np.array(list1)
    return x
```

```python
result=list()
for i in range(len(df)):
    answer_list=list()
    ans_org_crop=df['org_crop'][i]
    answer_list.append(ans_org_crop)
    ans_org=df['org'][i]
    answer_list.append(ans_org)
    ans_crop=df['crop'][i]
    answer_list.append(ans_crop)
    predictions=max(answer_list,key=answer_list.count)
    result.append(predictions)
```

```python
df['vote']=result
df
```

```python
submit=pd.DataFrame(df['filename'])
submit['label']=df['vote']
submit
```

```python
submit.to_csv('F:\\AICUP\\upload_private\\vote_private.csv',index=False)
```
